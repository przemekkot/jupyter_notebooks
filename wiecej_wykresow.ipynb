```python
import pandas
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pandas.DataFrame(prices)
df
```

```python
df = pandas.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
df
```

```python
prices_calculation = df['price'].apply(lambda price: price/4)
df['price'] = prices_calculation
df
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(kind='line', linestyle='dashed', title="Price of goods (USD)", color='red')
plt.ylim(0, 2)
plt.grid(True)
```
